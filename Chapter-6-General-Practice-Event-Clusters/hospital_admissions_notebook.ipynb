### Step 2 - Compare Cosine Distances Between True and Pseudo Clusters

```python
%%time
# measure cosine distances between all cluster centroids

true_cluster_cosine_distances = measure_cosine_distance_between_cluster(all_ordered_cluster_centroids.head(100))
```

```python
%%time

# 1. simulate clusters using interevent time distribution (pseudo clusters)
# 2. measure cosine distances between all pseudo clusters

num_iets = 18
num_clusters_to_make = 1000

# generate pseudo clusters
pseudo_clusters = create_pseudo_clusters(joined_df, num_iets, num_clusters_to_make)
# measure cosine distances
pseudo_cluster_cosine_distances = measure_cosine_distance_between_cluster(pseudo_clusters)
```

```python
%%time

# compare the above 2 distributions of true vs pseudo clusters 

compare_cosine_difference_cdf(true_cluster_cosine_distances, pseudo_cluster_cosine_distances,stat='proportion')
```

### Step 3 - Creating a Cluster Dictionary based on a Cosine Distance Threshold

```python
%%time

# identify an appropriate cosine threshold by finding the first minima in the distribution

cut_off_cosine_distance = 0.02

plot_distribution_cosine_distances(true_cluster_cosine_distances, cut_off_cosine_distance)
```

```python
%%time

# using the cosine threshold from above, create a cluster dictionary

cosine_distance_threshold = 0.02

cluster_dictionary, closest_matched_clusters = make_cluster_dictionary(all_ordered_cluster_centroids, cosine_distance_threshold)
```

```python
%%time

# visualise dictionary and frequency of similar clusters

plot_cluster_dictionary(cluster_dictionary, closest_matched_clusters)
```
